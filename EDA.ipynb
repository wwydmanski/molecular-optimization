```python
contact_points = [i.split("_")[0] for i in data.columns if "_contact" in i]

print("Number of unique structures:", data['Title'].nunique())
print("Numer of contact points:", len(contact_points))
```

```python
groups = [i.replace("A52","{0}") for i in data.columns if "A52" in i]

def is_identical(name):
    group_data = data[[i.format(name) for i in groups]].values
    return (group_data == group_data[0]).all()

unique_points = [i for i in contact_points if not is_identical(i)]

print("Number of contact points with differing fingerprints:", 52)
```
